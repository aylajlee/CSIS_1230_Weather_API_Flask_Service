### **2) GET with Query Parameters**

We'll use JSONPlaceholder (test API): `https://jsonplaceholder.typicode.com/posts`


````python

import requests

URL = "https://jsonplaceholder.typicode.com/posts"

params = {"userId": 1}
response = requests.get(URL, params=params, timeout=10)
print("Status:", response.status_code)
print("Final URL:", response.url)
print("First 2 posts:",response.json()[:2])
````


### 3) Sending Headers
Headers add extra information (e.g., content type, auth).


````python
URL = "https://jsonplaceholder.typicode.com/posts"

headers = {"Accept": "application/json"}
r = requests.get(URL, params={"userId": 2}, headers=headers, timeout=10)
print("Status:", r.status_code)
print("Content-Type:", r.headers.get("Content-Type"))
print("Preview:", r.json()[:2])

#timeout = defines the maximum amount of time, in seconds, t
# that the client will wait for a response before raising a requests.exceptions.Timeout exception
````

### **3.1 Sending Headers with AUTH (e.g API-Key, Authorization, e.t.c)**

````python
import requests

# The base URL and API Key
BASE_URL = "https://reqres.in/api"
API_KEY = "reqres_773e95ff00494eef890137e7737234ff"    # Your API key

# The headers dictionary that will be used for ALL requests
headers = {
    "X-api-key": API_KEY  # This is the crucial line. The key must be 'X-api-key'
}

# 1. POST: Create a New User
print("1. Creating a new user...")
post_url = f"{BASE_URL}/users"
new_user_data = {
    "first_name": "Pauline", 
    "last_name": "Clarke",
    "email": "pauline.clarke@dundermifflin.com"
}

post_response = requests.post(post_url, json=new_user_data, headers=headers)
print(f"POST Status Code: {post_response.status_code}")
print(f"POST Response: {post_response.json()}\n")
````

````python
# Extract the ID of the newly created user from the response
new_user_id = post_response.json().get('id')
print(f"New user created with ID: {new_user_id}\n")
````


### **4) POST — Create a Resource**


````python

URL = "https://jsonplaceholder.typicode.com/posts"

payload = {"title": "CSIS 1210", "body": "Week 4 demo", "userId": 200}
response = requests.post(URL, json=payload, timeout=10)

print("Status:", response.status_code)
print("Response JSON:", response.json())
````


### **5) PUT — Replace a Resource**


````python
URL = "https://jsonplaceholder.typicode.com/posts"

target = f"{URL}/1"
updated = {"id": 1, "title": "Updated title", "body": "Updated body", "userId": 1}
r = requests.put(target, json=updated, timeout=10)
print("Status:", r.status_code)
print("Response JSON:", r.json())
````


### **6) DELETE — Remove a Resource**


````python

r = requests.delete(target, timeout=10)
print("Status:", r.status_code)
try:
    print("Body:", r.json())
except Exception:
    print("No JSON body.")
````


### **7) Error Handling Pattern**


````python

def safe_get(url, **kwargs):
    try:
        r = requests.get(url, timeout=10, **kwargs)
        if r.ok:
            return r.json()
        else:
            print("Failed:", r.status_code)
            return None
    except requests.exceptions.RequestException as e:
        print("Network error:", e)
        return None

print(safe_get("https://jsonplaceholder.typicode.com/posts/1"))
print(safe_get("https://jsonplaceholder.typicode.com/badpath"))
````
