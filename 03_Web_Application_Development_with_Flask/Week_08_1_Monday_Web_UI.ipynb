
**Issues with this approach:**

- Difficult to maintain and scale

- No code reuse across pages

- Hard to collaborate with frontend developers

### **2. What is a Web UI?**

- Web UI = *User Interface built with HTML/CSS* that allows users to **see** data and **interact** with the backend.

- Flask provides the **bridge** between server logic (Python) and templates (HTML).

### 2.1. Introduction to Templating Engines

A template engine is a system that allows you to generate HTML with dynamic content using special syntax and placeholders.

***Flask*** uses **Jinja2**:

- Powerful, secure, and widely used

- Allows logic in templates (loops, conditionals)

- Supports template inheritance

- Automatically escapes HTML to prevent security issues

### 2.2. Jinja2 Template Basics
Flask uses the Jinja2 template engine.
Templates live inside a folder named **`templates/`** in your project directory.

### Example structure
```
weather_app/
 ├── app.py
 ├── templates/           
 │    └── base.html                (Parent)
 │        ├── index.html           (Child)
 │        ├── tasks.html           (Child)
 │        └── add_tasks.html       (Child)
 └── static/ (optional for CSS, JS)
```

### 2.3. The MVC (Model-View-Controller) Pattern

``Model (Data) → Controller (Flask Routes) → View (Jinja2 Templates)``

**In our context:**

- **Model:** Python data structures (lists, dictionaries)

- **Controller:** Flask route functions

- **View:** Jinja2 HTML templates



### 2.4 Jinja2 Template Syntax
| Syntax             | Purpose          | Example                        |
|--------------------|------------------|--------------------------------|
| `{{ variable }}`    | Output/Display     | `{{ task.title }}`              |
| `{% tag %}`          | Logic/Control       | `{% for item in list %}`         |
| `{# comment #}`      | Comments            | `{# This won't render #}`         |
| `{% extends %}`      | Inheritance          | `{% extends "base.html" %}`        |
| `{% block %}`         | Content blocks       | `{% block content %}{% endblock %}` |

### **3 Example – Render Data from Server**

```python
# app.py
from flask import Flask, render_template

app = Flask(__name__)

# Flask route controller
@app.route('/')
def home():
  cities = ["Chicago", "Austin", "New York"]
  return render_template('index.html', title="Weather Dashboard", cities=cities)

@app.route('/tasks')
def tasks():
  tasks = [{"title": "Buy groceries", "done": False},
           {"title": "Read book", "done": True}]
  return render_template('tasks.html', title="Task List", tasks=tasks)
```


**index.html**

````python
{% extends 'base.html' %}
{% block content %}
<h2>{{ title }}</h2>
<ul>
  {% for c in cities %}
    <li>{{ c }}</li>
  {% endfor %}
</ul>
{% endblock %}
````

**tasks.html**

```python 
{% extends 'base.html' %}
# In Jinja2 template (View) 
{% for task in tasks %}
    <div class="{{ 'completed' if task.completed else 'pending' }}">
        {{ task.title }}
    </div>
{% endfor %}
```

### **4 Handling Form Input**
Forms let users send data to the server (via POST).

### HTML form
```html
<form action="/submit" method="POST">
  <label>City Name:</label>
  <input type="text" name="city">
  <button type="submit">Search</button>
</form>
```

### Route to handle the form
```python
from flask import request

@app.route('/submit', methods=['POST'])
def submit():
  city = request.form.get('city')
  return render_template('result.html', city=city)
```

### **5 Good Practices**
| Do ✅ | Avoid 🚫 |
|---|---|
| Keep templates in `templates/` folder | Returning HTML inside Python strings for complex pages |
| Use `url_for()` to build links | Hard-coding URLs like `/submit` in multiple files |
| Validate form inputs before processing | Trusting user input blindly |
| Separate HTML layout (`base.html`) from content pages | Copy-pasting the same HTML in every template |

### **6 Common Mistakes**

- Forgetting `templates/` folder → `TemplateNotFound` error.

- Mixing tabs/spaces in Jinja loops.

- Returning raw strings instead of `render_template()` → loses context.

- Using wrong variable names (case-sensitive in Jinja!).
