### **1. The "Why": Virtual Environments and Dependencies**

**The Problem:** The "It Works on My Machine" Nightmare

Imagine you have two Python projects on your computer:

***Project A*** needs an older version of a library (e.g., Flask 2.1.0).

***Project B*** needs a new feature from a newer version of the same library (e.g., Flask 3.1.0).

Python can only have one version of a package installed globally at a time. If you upgrade Flask for ***Project B***, you might break ***Project A***. This is a classic dependency conflict .


### The Solution: Virtual Environments

A virtual environment is an isolated workspace for your project. It has its own:

- Python interpreter (a copy)

- Installed packages (```pip``` list)

What you install in one virtual environment **does not affect** other projects or the global system. This allows every project to have its own set of compatible dependencies .

**The Blueprint:** The ***requirements.txt*** File

The requirements.txt file is a simple text file that lists all the Python packages your project depends on, along with their specific versions .

### **2.  The "How": Step-by-Step Project Setup Guide**

Let's put theory into practice by setting up a new Flask project.

### Step 1: Create Your Project Directory

First, organize your files by creating a dedicated folder for the project.

**Create and navigate into your new project folder**

```mkdir my-flask-project```

```cd my-flask-project```

### Step 2: Create a Virtual Environment

Inside your project folder, create a virtual environment. It's best practice to name it simply ```.venv.```

``python3 -m venv .venv``

This command creates a new folder ``(.venv)`` containing a clean, isolated Python environment 


### Step 3: Activate the Virtual Environment

Before installing anything, you must activate the environment. 

This tells your terminal to use the Python and ```pip``` inside the ```.venv``` folder.

- **On macOS/Linux or Windows CMD:**

    `source .venv/bin/activate`

- **On Windows Powershell or Visual Studio:**

    ```.venv\Scripts\Activate.ps1```

### Step 4: Install Flask and Dependencies

With the environment active, install Flask. **``pip``** will automatically install Flask's core dependencies (Werkzeug, Jinja2, etc.)

```pip install Flask```

```pip install requests```

### Step 5: Generate Your requirements.txt File

Now, let's snapshot your project's dependencies. 

The **``pip freeze``** command lists all installed packages in the current environment.

`pip freeze > requirements.txt`

This creates the requirements.txt file. If you open it, you'll see Flask and all its dependencies with precise version numbers, which is crucial for consistency

**Note:** If you are just starting the project without having installed any dependencies, instead of running pip install manually for each dependency, you can add them in your requirements.txt (one dependency on each line) and install them into your virtual environment by running 

```pip install -r requirements.txt```


**Note:**

- Do not push **.venv** to github!

- Create a ``.gitignore`` file in your project folder

- add ```.venv ``` into the file
