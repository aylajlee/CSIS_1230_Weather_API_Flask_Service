### **What Are Aggregate Functions?** 

**Simple Definition:**  

Aggregate functions perform calculations on multiple rows and return a single result.  

**Example:**
  
Think of your grade report:  
- `COUNT`: How many assignments did you complete?  
- `SUM`: Total points earned  
- `AVG`: Your average grade  
- `MIN`: Your lowest score  
- `MAX`: Your highest score  

---

### **The 5 Main Aggregate Functions**  

| Function  | What It Does               | Example                         |
|-----------|----------------------------|---------------------------------|
| `COUNT()` | Counts how many rows        | How many tasks?                 |
| `SUM()`   | Adds up values             | Total points earned             |
| `AVG()`   | Calculates average         | Average task completion time    |
| `MIN()`   | Finds smallest value       | Earliest due date               |
| `MAX()`   | Finds largest value        | Latest due date                 |

### **1. COUNT Function**  

**Count All Tasks:**  
```sql
-- How many tasks do we have in total?
SELECT COUNT(*) AS total_tasks 
FROM tasks;


**Count Specific Tasks:**

````python
-- How many completed tasks?
SELECT COUNT(*) AS completed_tasks 
FROM tasks 
WHERE completed = true;

-- How many tasks have due dates?
SELECT COUNT(due_date) AS tasks_with_due_dates 
FROM tasks;
````

### **2. SUM and AVG Functions**

First, let’s add some data first to demonstrate this

````python
-- Add an 'estimated_hours' column to tasks table
ALTER TABLE tasks ADD COLUMN estimated_hours INTEGER;

-- Update some sample values
UPDATE tasks SET estimated_hours = 3 WHERE task_id = 1; -- 교수님 창에는 왜 오른쪽 화면이 보이는 거지? -> 6ms AffectedRows: 1 
UPDATE tasks SET estimated_hours = 5 WHERE task_id = 2;
UPDATE tasks SET estimated_hours = 2 WHERE task_id = 3;
UPDATE tasks SET estimated_hours = 8 WHERE task_id = 4;
````

````python
-- Total estimated hours for all tasks
SELECT SUM(estimated_hours) AS total_hours 
FROM tasks;

-- Average estimated hours per task
SELECT AVG(estimated_hours) AS average_hours 
FROM tasks;

-- Round the average to 2 decimal places
SELECT ROUND(AVG(estimated_hours), 2) AS average_hours 
FROM tasks;
````

### **3. MIN and MAX Functions**

**Find Date Ranges:**

````python
-- Earliest and latest due dates
SELECT 
    MIN(due_date) AS earliest_due_date,
    MAX(due_date) AS latest_due_date
FROM tasks 
WHERE due_date IS NOT NULL;
````

**Find Task Extremes:**

````python
-- Shortest and longest estimated tasks
SELECT 
    MIN(estimated_hours) AS shortest_task,
    MAX(estimated_hours) AS longest_task
FROM tasks 
WHERE estimated_hours IS NOT NULL;
````

### **4. The Magic of GROUP BY**

**What GROUP BY Does:**

Groups rows that have the same values into summary rows

**Intuition:**

Think of sorting laundry:

- All socks together → count socks

- All shirts together → count shirts

- All pants together → count pants

**Count Tasks by User:**

````python
-- How many tasks does each user have?
SELECT 
    user_id,
    COUNT(*) AS task_count
FROM tasks
GROUP BY user_id;
````

### GROUP BY in Action

**Tasks by Completion Status:**

````python
-- How many completed vs incomplete tasks?
SELECT 
    completed,
    COUNT(*) AS count
FROM tasks
GROUP BY completed;
````

**Tasks by User and Status:**

````python
-- Breakdown by user AND completion status
SELECT 
    user_id,
    completed,
    COUNT(*) AS task_count
FROM tasks
GROUP BY user_id, completed
ORDER BY user_id, completed;
````

### **5. Common Mistakes & Solutions**

**Mistake 1: Forgetting GROUP BY**

````python
-- ❌ WRONG: Which user_id? There are multiple!
SELECT user_id, COUNT(*) 
FROM tasks;

-- ✅ CORRECT: Group by user_id
SELECT user_id, COUNT(*) 
FROM tasks 
GROUP BY user_id;
````

**Mistake 2: Using WHERE instead of HAVING**

````python
-- ❌ WRONG: Can't use aggregate in WHERE
SELECT user_id, COUNT(*) 
FROM tasks 
WHERE COUNT(*) > 1 
GROUP BY user_id;

-- ✅ CORRECT: Use HAVING for aggregates
SELECT user_id, COUNT(*) 
FROM tasks 
GROUP BY user_id 
HAVING COUNT(*) > 1;
````

**Mistake 3: Including non-aggregated columns**

````python
-- ❌ WRONG: title isn't in GROUP BY
SELECT user_id, title, COUNT(*) 
FROM tasks 
GROUP BY user_id;

-- ✅ CORRECT: Either group by title or remove it
SELECT user_id, COUNT(*) 
FROM tasks 
GROUP BY user_id;
````

### **6. Aggregate Function Cheat Sheet**

**Basic Syntax:**

````python
SELECT 
    column1,
    AGGREGATE_FUNCTION(column2)
FROM table
WHERE conditions
GROUP BY column1
HAVING aggregate_conditions
ORDER BY column1;
````
