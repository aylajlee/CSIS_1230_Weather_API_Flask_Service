## **1. Introduction to SQL CRUD**

### What is SQL?

SQL (Structured Query Language) is the standard language for communicating with relational databases.

**The CRUD Operations:**

| Operation  | SQL Command | Purpose                     | HTTP Equivalent |
|------------|-------------|-----------------------------|-----------------|
| **Create** | `INSERT`    | Add new records             | POST            |
| **Read**   | `SELECT`    | Retrieve data               | GET             |
| **Update** | `UPDATE`    | Modify existing records     | PUT             |
| **Delete** | `DELETE`    | Remove records              | DELETE          |


## **2. CREATE: The INSERT Statement**

**Basic Syntax**
```sql
INSERT INTO table_name (column1, column2, column3, ...)
VALUES (value1, value2, value3, ...);
```

### Practical Examples

**2.1. Insert a New User:**


````python
INSERT INTO users (username, email, password_hash)
VALUES ('alice_dev', 'alice@email.com', 'hashed_password_123');
````

**2.2 Insert a Task:**

````python
INSERT INTO tasks (title, description, user_id, due_date)
VALUES ('Learn SQL', 'Practice CRUD operations', 1, '2024-03-15');
````

**2.3. Insert Multiple Tasks at Once:**

````python
INSERT INTO tasks (title, description, user_id) VALUES
('Build Flask App', 'Connect database to web app', 1),
('Study for Exam', 'Review chapters 1-5', 1),
('Team Project', 'Meet with group on Friday', 2);
````

**⚠️ Important Notes:**

- Auto-increment columns (like user_id, task_id) are automatically filled

- Required fields (NOT NULL) must be included

- Order matters - column order must match value order

## **3. READ: The SELECT Statement**

**Basic Syntax**

```sql
    SELECT column1, column2, ...
    FROM table_name
    WHERE condition
    ORDER BY column_name;
```

### Practical Examples

**3.1. Select All Columns:**

````python
SELECT * FROM users;
````

**3.2. Select Specific Columns:**

````python
SELECT username, email FROM users;
````

**3.3. Select with Filtering (WHERE):**

````python
-- Get completed tasks
SELECT title, description FROM tasks WHERE completed = true;

-- Get tasks due this week
SELECT title, due_date FROM tasks 
WHERE due_date BETWEEN '2024-03-10' AND '2024-03-17';

-- Get tasks for specific user
SELECT * FROM tasks WHERE user_id = 1;
````

**3.4. Select with Sorting:**

````python
-- Newest tasks first
SELECT title, created_at FROM tasks 
ORDER BY created_at DESC;

-- Alphabetical order
SELECT title FROM tasks 
ORDER BY title ASC;
````

**3.5. Select with Limits:**

````python
-- Get first 5 tasks
SELECT title FROM tasks LIMIT 5;

-- Get next 5 tasks (pagination)
SELECT title FROM tasks LIMIT 5 OFFSET 5;
````

## **4. UPDATE: Modifying Data**

**Basic Syntax**

````python
UPDATE table_name
SET column1 = value1, column2 = value2, ...
WHERE condition;
````

### ⚠️ CRITICAL SAFETY WARNING:
**ALWAYS use WHERE clause with UPDATE!**

````python
-- 🚨 DANGEROUS: Updates EVERY task in the table!
UPDATE tasks SET completed = true;

-- ✅ SAFE: Updates only specific tasks
UPDATE tasks SET completed = true WHERE task_id = 5;
````

### Practical Examples

**4.1. Mark Task as Complete:**

````python
UPDATE tasks 
SET completed = true 
WHERE task_id = 3;
````

**4.2. Update Multiple Fields:**

````python
UPDATE tasks 
SET title = 'Learn Advanced SQL', 
    description = 'Practice JOIN operations',
    due_date = '2024-03-20'
WHERE task_id = 5;
````

## **5. DELETE: Removing Data**

**Basic Syntax**

```sql
DELETE FROM table_name
WHERE condition;
```

### ⚠️ CRITICAL SAFETY WARNING:
**ALWAYS use WHERE clause with DELETE!**

````python
-- 🚨 CATASTROPHIC: Deletes EVERY task!
DELETE FROM tasks;

-- ✅ SAFE: Deletes only specific task
DELETE FROM tasks WHERE task_id = 2;
````

### Practical Examples

**1. Delete Specific Task:**

````python
DELETE FROM tasks WHERE task_id = 7;
````

**2. Delete Completed Tasks:**

````python
DELETE FROM tasks 
WHERE completed = true AND user_id = 1;
````

**3. Delete Old Tasks:**

````python
DELETE FROM tasks 
WHERE created_at < '2024-01-01';
````

## SUMMARY

**Basic CRUD Patterns:**

```sql

-- CREATE
INSERT INTO table (col1, col2) VALUES (val1, val2);

-- READ  
SELECT col1, col2 FROM table WHERE condition;

-- UPDATE
UPDATE table SET col1 = val1 WHERE condition;

-- DELETE
DELETE FROM table WHERE condition;
```
