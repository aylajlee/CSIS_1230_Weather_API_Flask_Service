### **3. Database Building Blocks**

### 3.1 Tables - The Containers

Think of tables as spreadsheets or Excel sheets

**Users Table**

| user_id | username    | email             |
|---------|-------------|-------------------|
| 1       | john_doe    | john@email.com    |
| 2       | jane_smith  | jane@email.com    |


**Columns - The Attributes**

- user_id (Integer, Primary Key)

- username (String, Unique)

- email (String, Unique)

**Rows - The Records**

- Each row represents one complete set of information




### 3.2 PostgreSQL Data Types

### Common Data Types

| Category   | Types                        | Description          |
|-------------|------------------------------|----------------------|
| **Text**    | `VARCHAR(n)`, `TEXT`         | Strings of text      |
| **Numeric** | `INTEGER`, `DECIMAL`, `SERIAL` | Numbers             |
| **Boolean** | `BOOLEAN`                    | True/False           |
| **Date/Time** | `DATE`, `TIMESTAMP`         | Dates and times      |
| **Special** | `UUID`                       | Unique identifiers   |


**Examples:**

````python
-- Creating a table with different data types
CREATE TABLE users (
    user_id SERIAL PRIMARY KEY,
    username VARCHAR(50) UNIQUE NOT NULL,
    email VARCHAR(100) UNIQUE NOT NULL,
    age INTEGER,
    is_active BOOLEAN DEFAULT true,
    created_at TIMESTAMP DEFAULT NOW()
);
````

### **5. Primary Keys & Foreign Keys**

### Primary Key (PK)

- Unique identifier for each row in a table

- Cannot be NULL

- Must be unique across the entire table

- Examples: user_id, task_id

### Foreign Key (FK)

- Reference to a primary key in another table

- Creates the relationship between tables

- Ensures **referential integrity**

````python
CREATE TABLE tasks (
    task_id SERIAL PRIMARY KEY,
    title VARCHAR(100) NOT NULL,
    user_id INTEGER REFERENCES users(user_id)  --> Foreign Key
);
````

### **6. Database Normalization**

**What is Normalization?**

The process of organizing data to ***reduce redundancy*** and improve ***data integrity***.

### 6.1 First Normal Form (1NF)

1. Each column contains atomic values (no lists/arrays)

2. Each record needs to be unique

**Before 1NF:**

tasks: "Learn Python, Study Math, Build Project"


| user_id  | task                                                | 
|----------|-----------------------------------------------------|
| 1        | Learn Python, Study Math, Build Project             |


**After 1NF:**

Task 1: "Learn Python"

Task 2: "Study Math" 

Task 3: "Build Project"

| user_id  | task                |      
|----------|---------------------|
| 1        | Learn Python        |
| 1        | Study Math          |
| 1        | Build Project       |


### 6.2 Second Normal Form (2NF)

1. Meets 1NF requirements

2. All non-key columns depend on the entire primary key

**Violation Example:**

❌ Order_Items table


| order_id | product_id | product_name   | category_name | quantity | unit_price |
|----------|------------|----------------|---------------|----------|------------|
| 1        | 101        | Python Book    | Education     | 2        | 29.99      |
| 1        | 102        | Mouse          | Electronics   | 1        | 19.99      |
| 2        | 101        | Python Book    | Education     | 1        | 29.99      |



Here, ``product_name ``depends only on ``product_id``, not the full primary key which is ``(order_id, product_id)`` .

- If we update a product name, we have to update multiple rows

- Data redundancy: "Python Book" appears multiple times

**Correction: Split the table into 2.**

***Products Table***

| product_id | product_name   | category_name | unit_price |
|------------|----------------|---------------|------------|
| 101        | Python Book    | Education     | 29.99      |
| 102        | Mouse          | Electronics   | 19.99      |

***Order_Items Table***


| order_id | product_id | quantity |
|----------|------------|----------|
| 1        | 101        | 2        |
| 1        | 102        | 1        |
| 2        | 101        | 1        |

### **8. Best Practices Summary**

### Naming Conventions

- Use snake_case for table and column names

- Table names: plural (users, tasks)

- Primary key: table_id (user_id, task_id)

- Foreign key: referenced_table_id (user_id)

### **9. Design Principles**

- Every table should have a primary key

- Use appropriate data types

- Establish proper relationships with foreign keys

- Apply normalization to reduce redundancy

- Add indexes for frequently queried columns
