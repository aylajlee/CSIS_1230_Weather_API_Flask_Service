```python
class Course:
    course_count = 0 
    
    # 1. SOME instance attributes
    # 2. ONE class attribute 
    def __init__(self, course_code, course_title, max_students):
        # instance 
        self.course_code = course_code
        self.course_title = course_title
        self.max_students = max_students
        Course.course_count += 1 
        # AFTER making a add_course()
        self.roster = [] # a list to store the roster 
        
    def add_course(self, student_name):
        if len(self.roster) < self.max_students:
            self.roster.append(student_name) # list # BUT, checking should be required!!
            print(f"Student {student_name} has been added to {self.course_code}.")
        else: 
            print(f"Cannot add student: course {self.course_code} is full.")
    
    def __str__(self):
        # [Submission]
        # return f"{self.course_code} - {self.course_title} ({self.course_count}/{self.max_students} students)"
        # [Professor's feedback]
        # current roster size, like " (len(self.roster)/max_students)"
        return f"{self.course_code} - {self.course_title} ({len(self.roster)}/{self.max_students} students)"
    
    
    def __repr__(self):
        return f"Course('{self.course_code}','{self.course_title}', {self.max_students})"
    
a_course = Course("CSIS1230", "Programming for Everyone", 3)
b_course = Course("CSIS1210", "Programming for Everyone", 5)

a_course.add_course("Alice Smith")
a_course.add_course("Juyeon")
a_course.add_course("Linnea")
a_course.add_course("Derrick")
a_course.add_course("Shyam")

# [The parts I missed]: Create TWO Course instances, so I should have created these 

# [Professor's feedback]
# only one course is printed/tested, 
# and the class attribute(course_count) isn't explicitly printed as required

b_course.add_course("Misa")

print(a_course)
print(repr(a_course))

print(b_course)
print(repr(b_course))

# [The parts I missed]: Print the class attribute tracking total courses 
# to verify it correctly counts both objects
print("Total courses:", Course.course_count)
```
