### **1) Intuition — why do developers use polymorphism & ABCs?**

**As a developer (Creator):** you want others to implement a small set of methods so your code can work with their classes **without changes**.

**As a developer (User):** you want to implement a small interface and immediately plug your class into someone else's system.

**Everyday analogies**

- **Remote control**: one set of buttons works for TVs, speakers, or projectors.

- **Drawing shapes**: `draw()` works for `Circle`, `Square`, or `Triangle`.

- **Payments**: `pay(amount)` works for different payment types.

Polymorphism keeps **client code** stable while allowing different behaviors underneath.



### **2) Polymorphism in practice (basic inheritance + overriding)**

- Shape defines what method is needed (area), but not how.

- Circle provides the implementation.


````python

class Shape:
    def area(self) -> float:
        return 0.0

class Circle(Shape):
    def __init__(self, radius: float):
        self.radius = radius

    def area(self) -> float:
        return 3.14 * self.radius ** 2

class Square(Shape):
    def __init__(self, side: float):
        self.side = side

    def area(self) -> float:
        return self.side ** 2

def print_areas(shapes: list[Shape]) -> None:
    for s in shapes:
        print("Area:", s.area())

print_areas([Circle(5), Square(3), Shape()])

````


### **3) Abstract Base Classes (ABCs) — how and when to use them**

An abstract class is a class that cannot be instantiated directly and is meant to serve as a template for other classes.


**What**: An ABC defines **required methods** (no implementation) that subclasses **must** implement.

**Why**:

- Communicates a clear **contract** to students/teammates.

- Prevents instantiating incomplete classes (early error).

- Lets you write code that accepts *any* future implementation that follows the contract.

**How** (with Python’s `abc` module):

1. Subclass `ABC`.

2. Mark required methods with `@abstractmethod`.

3. Subclasses must implement all abstract methods before you can create objects.


````python

from abc import ABC, abstractmethod

class Animal(ABC):
    
    def can_swim(self): # concrete method
        return False
    
    @abstractmethod
    def speak(self) -> str:
        """Return the sound this animal makes."""
        ...

class Dog(Animal):
    def speak(self) -> str:
        return "Woof!"

# Try it:
d = Dog()
print(d.speak())
### CAUTION: we have to use this method because of the above function like Animal(ABC)


# Uncomment to see the error:
# a = Animal()  # TypeError: Can't instantiate abstract class
````

### **4) Gentle, relatable scenarios**

### 4.1) Remote Control
We want the same buttons to work across devices.


````python

from abc import ABC, abstractmethod

class RemoteControl(ABC):
    @abstractmethod
    def switch_on(self) -> None: ...
    @abstractmethod
    def switch_off(self) -> None: ...

class TV(RemoteControl):
    def switch_on(self) -> None: print("TV ON")
    def switch_off(self) -> None: print("TV OFF")

class Speaker(RemoteControl):
    def switch_on(self) -> None: print("Speaker ON")
    def switch_off(self) -> None: print("Speaker OFF")

def use_remote(device: RemoteControl):
    device.switch_on()
    device.switch_off()

use_remote(TV())
use_remote(Speaker())
````


### **4.2) Payments**

Abstract classes help us to implement safe polymorphism 

- One checkout function; multiple ways to pay.

- As long as any processor class implements pay(), your code works—no need to know the concrete class.


````python

from abc import ABC, abstractmethod

class PaymentProcessor(ABC):
    @abstractmethod
    def pay(self, amount: float) -> str: ...

class CreditCard(PaymentProcessor):
    def pay(self, amount: float) -> str:
        return f"Paid ${amount:.2f} with credit card."

class Cash(PaymentProcessor):
    def pay(self, amount: float) -> str:
        return f"Paid ${amount:.2f} in cash."

def checkout(processor: PaymentProcessor, amount: float) -> None:
    print(processor.pay(amount))

checkout(CreditCard(), 24.0)
checkout(Cash(), 9.5)
````

### If it's still hazy, think of ABCs as non-instantiable “manager classes” that say what tasks all their team members must implement. 

### Any subgroup missing a required method cannot be hired—avoiding future chaos!


### **4.3) Good Habits**

- Keep ABC **small** and focused (1–3 methods).

- Add **docstrings** to explain what each method should do.

- Keep child method **signatures the same** as the ABC (same name/params/return).

- Prefer **clear names** (`pay`, `draw`, `play`) that describe the action.

- Write tests against the **interface**, not specific classes.



### **5) Common Pitfalls**

- **Instantiating an ABC** or forgetting to implement a required method → `TypeError`.

- **Changing method signatures** in subclasses (e.g., adding extra parameters) → breaks substitutability.

- **Too many layers** of inheritance → hard to follow; keep hierarchies shallow.

- **Overusing ABCs** when a simple base class or even a function would do.
