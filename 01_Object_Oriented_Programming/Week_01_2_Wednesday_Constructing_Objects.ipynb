### **1. Constructors (`__init__`)**

- Automatically called when an object is created

- First parameter is always `self`

- Used to initialize instance attributes

````python

class Student:
    def __init__(self, name, student_id):
        self.name = name                           # instance attribute
        self.student_id = student_id

alice = Student("Alice Chen", "S12345")
print(alice.name, alice.student_id)                # Creating an instance
````

### **2. Object Representation: `__str__` vs `__repr__`**

- `__str__`: user-friendly representation

- `__repr__`: developer-friendly, ideally re-creatable


| Method   | Purpose                    | Audience   | Called by              |
|----------|----------------------------|------------|------------------------|
| `__str__`  | Human-readable output       | End-users  | `str()`, `print()`     |
| `__repr__` | Unambiguous debugging output | Developers | `repr()`, console      |


````python

class Student:
    def __init__(self, name, student_id):
        self.name = name
        self.student_id = student_id

    def __str__(self):
        return f"Student: {self.name} ({self.student_id})"

    def __repr__(self):
        return f"Student(name='{self.name}', student_id='{self.student_id}')"

alice = Student("Alice Chen", "S12345")
print(alice)
print(repr(alice))
````

### **3. Instance vs Class Attributes**

- **Instance attributes**: 
    - unique per object (```self.attribute```)
    
    - Defined in __init__ or methods
    

- **Class attributes**: 
    - Shared across all instances
    - Defined directly under class


````python

class Student:
    school = "North Park University"  # class attribute

    def __init__(self, name):
        self.name = name              # instance attribute

alice = Student("Alice")
bob = Student("Bob")

print(alice.school)                   # Adopted by all instances
alice.school = "Stanford"

print(Student.school)
print(alice.school)
print(bob.school)
````

### **4. Stateful Classes**

- Maintain internal state between method calls

- State changes over time

````python

class BankAccount:
    def __init__(self, balance=0):
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount

account = BankAccount(100)
account.deposit(50)
account.withdraw(30)
print(account.balance)
````

### **5. Stateless Classes**

- No instance attributes

- Methods depend only on input arguments

- Often implemented using `@staticmethod`

````python

class MathUtils:
    @staticmethod
    def add(a, b):
        return a + b

    @staticmethod
    def multiply(a, b):
        return a * b

# Usage  - no instance needed!
print(MathUtils.add(3, 5))
print(MathUtils.multiply(4, 6))
````

### 5.1 When to use stateless classes

- **Shared Utilities:** When you need reusable functions grouped logically

- **Constants/Configs:** For read-only settings shared across the app.

- **Functional Programming:** To avoid side effects.

- **Performance:** Avoid instance overhead when state isn't needed.


### 5.2 Static Methods

- Do not access instance (`self`) or class (`cls`) state

- Used for utility-style functionality

### 5.3 Class Methods

- Bound to the class, not an instance. 

- `cls` is the first parameter (not `self`)

- Often used as **factory methods** or to modify class-level state

- Cannot access instance-specific data (since they don’t take `self`)

````python
# Student Class with @classmethod

class Student:
    university = "Stanford"                                                # Class level attribute

    def __init__(self, name, age, major):
        self.name = name
        self.age = age
        self.major = major

    def describe(self):                                                    # Regular instance method
        return f"{self.name}, {self.age}, studying {self.major}"

    @classmethod
    def from_dict(cls, data):                                              # class method : Alternative constructor (from dictionary)
        return cls(data["name"], data["age"], data["major"])

    @classmethod                                                            
    def change_university(cls, new_university):                             # Modify class level attribute
        cls.university = new_university

student_data = {"name": "Alice", "age": 20, "major": "CS"}
s = Student.from_dict(student_data)
print(s.describe())
Student.change_university("North Park University")
print(Student.university)
````
